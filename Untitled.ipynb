```python
import torch 
import torch.nn as nn
import math
```

```python
import matplotlib.pyplot as plt
```

```python
G = torch.Tensor([[1,1],[3,2]])
phi = torch.tensor(math.pi / 2)
s = torch.sin(phi)
c = torch.cos(phi)
rot = torch.stack([torch.stack([c, -s]),
                   torch.stack([s, c])])
```

```python
G[1,:]
```

```python
x_rot = G @ rot.t()
```

```python
plt.scatter(y=G[1,:], x=G[0,:])
plt.scatter(y=x_rot[1,:],x=x_rot[0,:] )
```
